# Scatter Plot
---

```python
import matplotlib.pyplot as plt

# ข้อมูล ราคาหุ้น 
Date = [1,2,3,4,5,8,9,10,11,12,16,17,18,19,22,23,24,25,26]
Price = [750.46,784.5,770.22,722.81,703.87,667.85,672.32,686.86,675,690.26,692.98,699,
710,695.03,707.45,701.45,688.92,700.01, 708.58]

plt.scatter(Date, Price)
plt.xlabel("Date")
plt.ylabel("Price per unit")
plt.show()
```

# Line Plot
---

```python
import matplotlib.pyplot as plt

# ข้อมูล ราคาหุ้น 
Date = [1,2,3,4,5,8,9,10,11,12,16,17,18,19,22,23,24,25,26]
Price = [750.46,784.5,770.22,722.81,703.87,667.85,672.32,686.86,675,690.26,692.98,699,
710,695.03,707.45,701.45,688.92,700.01, 708.58]

Date_sort, Price_sort  = zip(*sorted(zip(Date, Price)))

plt.plot(Date_sort, Price_sort, 'b-*', linewidth=2, markersize=12)
plt.xlabel("Date")
plt.ylabel("Price per unit")
plt.show()
```

# Bar Plot

```python
import matplotlib.pyplot as plt
import numpy as np

Day = ['Thu', 'Fri', 'Sat', 'Sun']
Tables = [62, 19, 87, 76]

plt.bar(Day,Tables)
plt.show()
```

# Pie Plot
---

```python
import matplotlib.pyplot as plt
import numpy as np

Day = ['Thu', 'Fri', 'Sat', 'Sun']
Tables = [62, 19, 87, 76]

plt.pie(Tables, labels= Day, autopct='%1.2f%%')
plt.show()
```

# Box Plot
---

```python
# เรียกใช้ Libraries Seaborn
import seaborn as sns

# Load ข้อมูล Tips
tips = sns.load_dataset("tips")

# ใช้คำสั่ง Box Plot แสดงค่า Tips มื้อกลางวัน และ ค่ำ 
sns.boxplot(x="time", y="tip", data=tips)
```

# Violin Plot
---

```python
# เรียกใช้ Libraries Seaborn
import seaborn as sns

# Load ข้อมูล Tips
tips = sns.load_dataset("tips")

# ใช้คำสั่ง Violin Plot แสดงการกระจายตัวค่า Tips มื้อกลางวัน และ ค่ำ 
sns.violinplot(x="time", y="tip", data=tips)
```
